# Kaggleのタイタニック問題

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
 
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
titanic_train = pd.read_csv('./titanic/train.csv')
titanic_test = pd.read_csv('./titanic/test.csv')
```

```python
# データサイズ
print('学習データ: {}'.format(titanic_train.shape))
print('評価データ: {}'.format(titanic_test.shape))
```

```python
titanic_train.head(n=10)
```

## 欠損データの確認

```python
pd.concat({'count': titanic_train.isnull().sum(), 'rate': 100*titanic_train.isnull().mean()}, axis=1).query('count > 0')
```

```python
pd.concat({'count': titanic_test.isnull().sum(), 'rate': 100*titanic_test.isnull().mean()}, axis=1).query('count > 0')
```

## 穴埋め

```python
titanic_train["Age"] = titanic_train["Age"].fillna(titanic_train["Age"].mode()[0])
titanic_train["Cabin"] = titanic_train["Cabin"].fillna(titanic_train["Cabin"].mode()[0])
titanic_train["Embarked"] = titanic_train["Embarked"].fillna(titanic_train["Embarked"].mode()[0])

titanic_test["Age"] = titanic_test["Age"].fillna(titanic_test["Age"].mode()[0])
titanic_test["Fare"] = titanic_test["Fare"].fillna(titanic_test["Fare"].mode()[0])
titanic_test["Cabin"] = titanic_test["Cabin"].fillna(titanic_test["Cabin"].mode()[0])
```

## カラム変更

```python
titanic_train = pd.get_dummies(titanic_train, columns=["Sex", "Embarked"])
titanic_test = pd.get_dummies(titanic_test, columns=['Sex', 'Embarked'])
```

```python
titanic_train.head()
```

## k-最近傍法で予測

```python
clf = KNeighborsClassifier(n_neighbors=1)

x_train = titanic_train[["Pclass", "Age", "SibSp", "Parch", "Fare", "Sex_female", "Sex_male", "Embarked_C", "Embarked_Q", "Embarked_S"]]
y_train = titanic_train["Survived"]

x_test = titanic_test[["Pclass", "Age", "SibSp", "Parch", "Fare", "Sex_female", "Sex_male", "Embarked_C", "Embarked_Q", "Embarked_S"]]

clf.fit(x_train, y_train)

# 予測結果
y_test = clf.predict(x_test)
```

## 解答ファイルの出力

```python
passenger_id = np.array(titanic_test["PassengerId"]).astype(int)
answer = pd.DataFrame(y_test, passenger_id, columns=["Survived"])
answer.to_csv("titanic_answer.csv", index_label=["PassengerId"])
```
